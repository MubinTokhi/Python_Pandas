```python
import pandas as pd
```

```python
macro_con = pd.read_csv('USDA_Macronutrients.csv')
micro_con = pd.read_csv('USDA_Micronutrients.csv')
```

```python
macro_con.head()
```

```python
micro_con.head()
```

```python
USDA = pd.merge(macro_con,micro_con,how = 'inner', on='ID')
USDA.head()
```

```python
USDA.dtypes
```

```python
USDA_DROP_NA =USDA.dropna(thresh = 10)
len(USDA_DROP_NA)
USDA_DROP_NA
```

```python
#For records with missing values for Sugar, Vitamin E and Vitamin D, replace missing values with mean value for the respective variable.
USDA['Sugar']= USDA['Sugar'].fillna(USDA['Sugar'].mean())
USDA['VitaminD'] = USDA['VitaminD'].fillna(USDA['VitaminD'].mean())
USDA['VitaminE'] = USDA['VitaminE'].fillna(USDA['VitaminE'].mean())
USDA
```

```python
USDA_clean = USDA.dropna(how = 'any')
len(USDA_clean)
USDA_clean
```

```python
# Which food has the highest sodium level?



USDA_clean['Sodium'].max()
# OR
USDA_MAX = USDA_clean.sort_values('Sodium',ascending = False)
USDA_MAX["Description"].head(1)
```
