# Pandas Hotel Data Project

# The Data

This data set contains booking information for a city hotel and a resort hotel, and includes information such as when the booking was made, length of stay, the number of adults, children, and/or babies, and the number of available parking spaces, among other things.

All personally identifying information has been removed from the data.

Acknowledgements
The data is originally from the article Hotel Booking Demand Datasets, written by Nuno Antonio, Ana Almeida, and Luis Nunes for Data in Brief, Volume 22, February 2019.


----------------------------

#### NOTE: Names, Emails, Phone Numbers, and Credit Card numbers in the data are synthetic and not real information from people. The hotel data is real from the publication listed above.

**the names of the people who had booked the most number children and babies for their stay**

```python
my_file['total_kids'] = my_file['babies'] + my_file['children']

my_file.sort_values('total_kids',ascending=False)[['name','adults','total_kids','babies','children']][:3]
```

**Top 3 most common area code in the phone numbers**

```python
my_file['phone-number'].str.split('-', expand = True)[0].value_counts()[:3]

# OR

my_file['phone-number'].apply(lambda num:num[:3]).value_counts()[:3]
```

**arrivals took place between the 1st and the 15th of the month**

```python
def arrivales(day):
    arrivals = 0 
    if (day>=1) & (day<=15):
        arrivals+=1
    return arrivals    

my_file['arrival_date_day_of_month'].apply(arrivales).sum()
```
